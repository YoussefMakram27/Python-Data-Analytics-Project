```python
# 7
data = np.array([150,220,90,300,180,45])

data_norm = (data - data.min()) / (data.max() - data.min())

print(data)
print(data_norm)
```

```python
# 8
data = np.array([150,220,90,300,180,45])

data_std = (data - data.mean()) / data.std()

print(data)
print(data_std)
```

```python
# 9
mat = np.array([
    [4,8,1,12],
    [7,3,9,5],
    [11,2,6,10],
    [15,0,14,13]
])

print(mat.mean(axis=1))
print(mat.sum(axis=1))

print('Max: ', mat.max())

row, col = np.unravel_index(mat.argmax(), mat.shape)
print('Max Position: ', row, col)
```
